```python
data=open('dataset.csv')
read=csv.reader(data)
transaction_list=[]

for r in read:
    transaction_list.append(r)
    
    
adjusted_list = []
for i in transaction_list:
    adjusted_list.append(i[0].split(','))
    
    
te = TransactionEncoder()
te_ary = te.fit(adjusted_list).transform(adjusted_list)
dataset = pd.DataFrame(te_ary, columns=te.columns_)
dataset
```
